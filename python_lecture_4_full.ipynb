```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(features,labels,test_size=0.2)
```

```python
model = svm.SVC()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```
